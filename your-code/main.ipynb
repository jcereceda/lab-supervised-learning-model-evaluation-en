## Regression Model Evaluation

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
"""
CRIM - per capita crime rate by town
ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
INDUS - proportion of non-retail business acres per town.
CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
NOX - nitric oxides concentration (parts per 10 million)
RM - average number of rooms per dwelling
AGE - proportion of owner-occupied units built prior to 1940
DIS - weighted distances to five Boston employment centres
RAD - index of accessibility to radial highways
TAX - full-value property-tax rate per $10,000
PTRATIO - pupil-teacher ratio by town
B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
LSTAT - % lower status of the population
MEDV - Median value of owner-occupied homes in $1000's"""
```

```python
data
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
# Your code here :
from sklearn.model_selection import train_test_split
seed = 1

X = data.drop('MEDV', axis=1)  # Todas las columnas (X) EXCEPTO Y
y = data['MEDV']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    random_state=seed
)
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

```python
# Your code here :
from sklearn.linear_model import LinearRegression

model_linear = LinearRegression()

model_linear.fit(X_train, y_train)
```

## 3. Calculate and print R-squared for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import mean_squared_error
import numpy as np

hw_test = model_linear.predict(X_test)
hw_train = model_linear.predict(X_train)

r2_test = model_linear.score(X_test, y_test)

r2_train = model_linear.score(X_train, y_train)

print(f"VALORES TEST")
print(f"R² (Coeficiente de Determinación): {r2_test:.4f}")
print(f"########################################")
print(f"VALORES TRAIN")
print(f"R² (Coeficiente de Determinación): {r2_train:.4f}")
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
# Your code here :
mse_test = mean_squared_error(y_test, hw_test)

mse_train = mean_squared_error(y_train, hw_train)

print(f"VALORES TEST")
print(f"MSE (Error Cuadrático Medio):     {mse_test:.4f}")
print(f"RMSE (Raíz del MSE):               {np.sqrt(mse_test):.4f}")
print(f"########################################")
print(f"VALORES TRAIN")
print(f"MSE (Error Cuadrático Medio):     {mse_train:.4f}")
print(f"RMSE (Raíz del MSE):               {np.sqrt(mse_train):.4f}")
```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
# Your code here :
actual = y.mean()

mae = actual - mse_test

print(f"MAE = {mae}")
```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
# Your code here :
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='species')

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                   test_size=0.2,
                                   random_state=seed,
                                   stratify=y)
```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
# Your code here :
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()

model.fit(X_train, y_train)


y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
# Your code here :
accuracy_test = accuracy_score(y_test, y_pred_test)

accuracy_train = accuracy_score(y_train, y_pred_train)

print(f"Accuracy (Exactitud) test: {accuracy_test*100:.2f}%")
print(f"Accuracy (Exactitud) train: {accuracy_train*100:.2f}%")
```

## 9. Calculate and print the balanced accuracy score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import balanced_accuracy_score

# Balanced accuracy para test y train
balanced_accuracy_test = balanced_accuracy_score(y_test, y_pred_test)
balanced_accuracy_train = balanced_accuracy_score(y_train, y_pred_train)

print(f"Balanced Accuracy test: {balanced_accuracy_test*100:.2f}%")
print(f"Balanced Accuracy train: {balanced_accuracy_train*100:.2f}%")
```

## 10. Calculate and print the precision score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import precision_score

# Precision para test y train
precision_test = precision_score(y_test, y_pred_test, average='macro')
precision_train = precision_score(y_train, y_pred_train, average='macro')
print(f"Precision test: {precision_test*100:.2f}%")
print(f"Precision train: {precision_train*100:.2f}%")
```

## 11. Calculate and print the recall score for both the training and the testing set.

```python
# Your code here :

from sklearn.metrics import recall_score

# Recall para test y train
recall_test = recall_score(y_test, y_pred_test, average='macro')
recall_train = recall_score(y_train, y_pred_train, average='macro')

print(f"Recall test: {recall_test*100:.2f}%")
print(f"Recall train: {recall_train*100:.2f}%")
```

## 12. Calculate and print the F1 score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import f1_score

# F1 score para test y train
f1_test = f1_score(y_test, y_pred_test, average='macro')
f1_train = f1_score(y_train, y_pred_train, average='macro')
print(f"F1 Score test: {f1_test*100:.2f}%")
print(f"F1 Score train: {f1_train*100:.2f}%")
```

## 13. Generate confusion matrices for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import confusion_matrix

# Matriz de confusión para el conjunto de test
cm_test = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix (Test):")
print(cm_test)

# Matriz de confusión para el conjunto de train
cm_train = confusion_matrix(y_train, y_pred_train)
print("\nConfusion Matrix (Train):")
print(cm_train)
```
